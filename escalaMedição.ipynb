```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('nominalData.csv')


coluna_alvo = 'Espécie'


plt.hist(df[coluna_alvo], bins='auto', alpha=0.7, color='brown', edgecolor='black')


plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.title(f'Histograma de {coluna_alvo}')


plt.show()

```

```python

df2 = pd.read_csv('ordinalData.csv')

coluna_alvo = 'Satisfacao'

plt.hist(df2[coluna_alvo], bins='auto', alpha=0.7, color='brown', edgecolor='black')

plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.title(f'Histograma de {coluna_alvo}')
plt.show()

contagem = df2['Satisfacao'].value_counts()

porcentagens = (contagem / len(df2)) * 100
porcentagens_inteiras = porcentagens.round(0).astype(int)
porcentagens_formatadas = porcentagens_inteiras.apply(lambda x: f'{x}%')
print(porcentagens_formatadas)
```

```python

df3 = pd.read_csv('continuousData.csv')
descricao = df3.describe()

coluna_salario = 'Salario'

plt.bar(df3.index, df3[coluna_salario], color='blue', alpha=0.7)

plt.xlabel('Índice')
plt.ylabel('Salário')
plt.title(f'Gráfico de Barras - Coluna {coluna_salario}')

plt.show()

coluna_salario = 'Salario'

plt.boxplot(df3[coluna_salario])

plt.xlabel('Coluna Salário')
plt.ylabel('Salário')
plt.title(f'Boxplot - Coluna {coluna_salario}')

plt.show()
```
